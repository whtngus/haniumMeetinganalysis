```python
## 발언자별   수업대화_과외수업 불러오기

f = open("C:/Users/hm/Desktop/회의데이터/20140109_수업대화_과외수/7CM00009.txt", 'rt', encoding='utf-16')
name = ""
talk = []
speak = {}

while(True):
    data = f.readline() 
    if not data: break
    
    if("<u who=" in data or "<s n=" in data):
        if "who" in data:
            name = data[8:10]
            if not (data[8:10] in speak):
                speak[name] = [] 

        temp = data.split(">")
        #speak[name] = speak[name] + [v[:-3] for v in temp if "/s"  in v or ("pause"  in v and v.find("<") != 0)]
        for v in temp:
            if "/s"  in v:
                speak[name].append(v[:-3])
            
            if "pause"  in v and v.find("<") != 0:
                speak[name].append(v[:-7])
            


print(speak)
```

```python
## 발언순서별   수업대화_과외수업 불러오기

f = open("C:/Users/hm/Desktop/회의데이터/20140109_수업대화_과외수/7CM00009.txt", 'rt', encoding='utf-16')
talk = []

while(True):
    data = f.readline() 
    if not data: break
    
    if("<s n=" in data):

        # 대화 추출
        temp = data.split(">")
        s = ""
        for v in temp:
            idx = v.find('<')
            if idx < 2: continue
            talk.append(v[:idx])


for t in talk:
    print(t)
```
